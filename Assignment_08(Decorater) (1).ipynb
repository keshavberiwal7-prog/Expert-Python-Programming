```python
# Create a decorator function
def decor(func):
  def wrapper():
    # Add any decoration by function
    print("Not start")
    func()

    print("Start")

# return the decorated function
  return wrapper

# basic syntax
@decor
def printer():
  print("Hello!")
  print("2Hello!")

# Directly calling the decorated function
printer()
```

```python
# Create a decorator function
def decorator_function(func):
  def wrapper():
    # Add any decoration by function
    print("Before the original function")
    func() # this is a original function
    print("After the original function")

  # Print the decorated function
  return wrapper

#Basic Syntax
@decorator_function
def original_function():
  print("Inside the original function")
  pass

# Directly calling the decorated function
original_function()
```

```python
# Modify the wrapper to accept *args and **kwargs.
# Create a Decorators with Arguments
def my_decorator(func):
  def wrapper(*args, **kwargs):
    print("Before the function is called")
    result = func(*args, **kwargs)
    # Explicitly print the result of the decorated function here
    print(f"Function result: {result}")
    print("After the function is called")
    return result
  return wrapper

@my_decorator
def add(a, b):
  return a + b

# Call the decorated function and assign its return value to a variable
result = add(5, 5)
```

#1. Logging Decorator (Fixed Arguments)

**Question:** Create a decorator that logs function name and parameters. The function should take exactly 2 arguments.


```python
def log_function(func):
  def wrapper(a, b):
    print(f"Callint {func.__name__} function, with argument {a}, {b}")
    result = func(a, b)
    print(f"{func.__name__} Result: {result}")
  return wrapper

@log_function
def multiply(a, b):
  return a * b

d = multiply(5, 5)
```

#2. Timer Decorator (No Arguments)

**Question :** Create a decorator that measures excution time of a function with no parameters.

```python
import time

def timer_decor(func):
  def wrapper():
    start = time.time()
    result = func()
    end = time.time()
    print(f"Excutation time {end - start:.4f} seconds.")
    return result
  return wrapper

@timer_decor
def do_work():
  time.sleep(3)
  print("Work Done")

do_work()
```

#3. Authorization Decorator (Single String Parameter)

**Question:** Create a decorator that checks if a user is "admin".

```python
def required_admin(func):
  def wrapper(user):
    if user != "Admin":
      print("Access denied")
      return
    return func(user)

  return wrapper

@required_admin
def access_dashboard(user):
  print(f"Welcome {user} :- Your dashboard accessed")


access_dashboard("guest")
access_dashboard("Admin")
```

#Returning Values

Decorators can also return values from the wrapped function:

```python
def double_result(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)
        return result*2
    return wrapper

@double_result
def get_number():
    return 10

print(get_number())
```
