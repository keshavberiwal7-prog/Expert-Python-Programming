Q1. Create an ElectricCar class that inherits from the Car class and includes an additional attribute, battery_size.

```python
class Car:
    def __init__(self, model,brand):
        self.model = model
        self.brand = brand
class type(Car):
    def __init__(self,model,brand,battery_size):
        super().__init__(model,brand)
        self.battery_size=battery_size
car1=type("Baleno","Maruti","90KW")
print(car1.brand,car1.model,car1.battery_size)
```

Q2.Define a Laptop class that inherits from the Computer class and includes an additional attribute, battery_life.


```python
class laptop:
    def __init__(self,brand,proccessor):
        self.brand=brand
        self.proccessor=proccessor
class battery(laptop):
    def __init__(self,brand,proccessor,battery_life):
        super().__init__(brand,proccessor)
        self.battery_life=battery_life
s1=battery("Lenovo","H-type","6hours")
print(s1.brand,s1.proccessor,s1.battery_life)
```

Q3.Create a Smartphone class and add another child class of your choice that inherits from an ElectronicDevice class. Include an additional attribute, operating_system.

```python
class smartphone():
    def __init__(self,brand,storage):
        self.brand=brand
        self.storage=storage
class device(smartphone):
    def __init__(self,brand,storage,os):
        super().__init__(brand,storage)
        self.os=os
x=device("Samsung Galaxy S25 Ultra","12gb","Android 13")
print(x.brand,x.storage,x.os)
```

Q4.Implement a Musician class and create another child class of your choice that extends an Artist class.

```python
class musician():
    def __init__(self,name,age):
        self.name=name
        self.age=age
class artist(musician):
    def __init__(self,name,age,play):
        super().__init__(name,age)
        self.play=play
x= artist("Zakhir Hussain",73,"Tabla")
print(x.name,x.age,x.play)
x2= artist("A.R Rahman",58,"Piano")
print(x2.name,x2.age,x2.play)
```

Q5.Define a FictionBook class and add another child class of your choice that inherits from the Book class. Include an additional attribute, genre.

```python
class fictionbook:
    def __init__(self,writer,name):
        self.writer=writer
        self.name=name
class book(fictionbook):
    def __init__(self,writer,name,genre):
        super().__init__(writer,name)
        self.genre=genre
s1=book("SPY×FAMILY1","Tatsuya Endo","slice-of-life")
print(s1.writer,s1.name,s1.genre)
```
