Question 1:
Create a Laptop class with a private attribute __price. Implement a method set_price(new_price) to update the price and get_price() to retrieve the price.


```python
class Laptop:
  def __init__(self, price):
    self.__price = price

  def set_price(self, new_price):
    if new_price > 0:
      self.__price = new_price

  def get_price(self):
    return self.__price
```

```python
L_price = Laptop(100)
L_price.set_price(200)
L_price.get_price()
```

Question 2:
Create a BankAccount class with a private attribute __balance. Implement methods to deposit, withdraw, and check balance while ensuring the balance cannot be modified directly.


```python
class BankAccount:
  def __init__(self, balance):
    self.__balance = balance

  def deposite(self, amount):
    if amount > 0:
      self.__balance += amount

  def withdraw(self, amount):
    if amount > 0 and amount <= self.__balance:
      self.__balance -= amount

  def get_balance(self):
    return self.__balance
```

```python
acc = BankAccount(100)
acc.deposite(200)
acc.withdraw(20)
acc.get_balance()
```

Question 3:
Create a Movie class with a private attribute __rating. Implement methods to set and get the rating, ensuring the rating is between 1 and 10.


```python
class Movie:
  def __init__(self, rating):
    self.__rating = rating

  def set_rating(self, new_rating):
    if new_rating > 0 and new_rating <= 10:
      self.__rating = new_rating

  def get_rating(self):
    return self.__rating
```

```python
RM = Movie(6)
RM.set_rating(9)
RM.get_rating()
```

Question 4:
Create a Book class with private attributes __title and __author. Implement a constructor to set these values and provide methods to retrieve them.

```python
class Book:
  def __init__(self, title, author):
    self.__title = title
    self.__author = author

  def set_title(self, new_title):
    self.__title = new_title

  def set_author(self, new_author):
    self.__author = new_author

  def get_title(self):
    return self.__title

  def get_author(self):
    return self.__author
```

```python
TA = Book("AA","BB")
TA.set_title("AI Impact")
TA.set_author("AKS")
```

Create a Car class with a private attribute __speed. Implement methods to increase speed, decrease speed, and get the current speed, ensuring speed does not go below 0.


```python
class Car:
  def __init__(self, speed):
    self.__speed = speed

  def set_increase(self, new_speed):
    if new_speed > 0:
      self.__speed += new_speed

  def set_decrease(self, new_speed):
    if new_speed > 0 and new_speed <= self.__speed:
      self.__speed -= new_speed

  def get_speed(self):
    return self.__speed
```

```python
CS = Car(50)
CS.set_increase(10)
CS.get_speed()
```

```python
CS = Car(50)
CS.set_decrease(10)
CS.get_speed()
```
